```r
library(tidyverse)
```

```r
# Load data
data <- read_csv("data/RaguseaMeatWashingSurveyResponsesRaw.csv", skip = 1)
names(data)<-str_replace_all(names(data), c(" " = "_" , "," = ""))
names(data) <- gsub(x = names(data), pattern = "\\?", replacement = "")
names(data) <- gsub(x = names(data), pattern = "\\'", replacement = "")

data
colnames(data)
```

```r
#filter to just USA and remove rows with all NAs
data <- filter(data, Prep_Meat!="Never" & Country_Residence=="United States of America", Houshold_size > 0)

head(data,16)
```

```r
# Create main dataset
data_main <- select(data, Houshold_size, Food_Safety_Important, Different_plates:Wash_Ground_Meat)

names(data_main) <- gsub(x = names(data_main), pattern = "Wash_", replacement = "")

# data_main <- pivot_longer(data_main,
#               cols = Pork:Ground_Meat,
#               names_to = "Meat_Type",
#               values_to = "Wash_Or_No_Wash"
#             )

# data_main <- filter(data_main, !is.na(Wash_Or_No_Wash))
# data_main <- mutate(data_main, Wash_Or_No_Wash = recode(Wash_Or_No_Wash, "Wash" = 1, "Dont Wash" = 0))

# colnames(data_main) <- c("Houshold_size", "Food_safety_importance", 
#                                  "Different_plates", "Thermometer",
#                                  "Not_washing_between_utensils", "Not_wash_between_cutting_boards", 
#                          "Leave_food_out_two_hours_or_more", "Meat_type", "Wash_Or_No_Wash")

colnames(data_main) <- c("Houshold_size", "Food_safety_importance", 
                                 "Different_plates", "Thermometer",
                                 "Different_utensil", "Different_cutting_boards", 
                         "Not_leaving_food_out", "Pork", "Poultry", "Beef", "Sheep_Goat", "Fish", "Ground_Meat")


data_main <- mutate(data_main, Houshold_size = ifelse (Houshold_size >= 6, 6, ceiling(Houshold_size)))

data_main <- replace(data_main, data_main == "Wash", "1") %>%
                replace(data_main == "Dont Wash", "0")

data_main <- mutate(data_main, Pork = as.numeric(Pork),
                                Poultry = as.numeric(Poultry),
                                Beef = as.numeric(Beef),
                                Sheep_Goat = as.numeric(Sheep_Goat),
                                Fish = as.numeric(Fish),
                                Ground_Meat = as.numeric(Ground_Meat))

#create the wash any column, 1 if respondant washes any type of meat
data_main <- mutate(data_main, Wash_Any = ifelse((Pork == 1 | 
                                                           Poultry == 1 |
                                                           Beef == 1 |
                                                           Sheep_Goat == 1 |
                                                           Fish == 1 |
                                                           Ground_Meat == 1 ), 1, 0))

# recode the food safety answers so that answering "Always" means food safe

data_main <- mutate(data_main, Different_utensil = replace(Different_utensil,
                                                            Different_utensil == "Never",
                                                           "always")) %>%
            mutate(Different_utensil = replace(Different_utensil,
                                                            Different_utensil == "Seldom",
                                                           "usually")) %>%
            mutate(Different_utensil = replace(Different_utensil,
                                                            Different_utensil == "Sometimes",
                                                           "about half the time")) %>%
            mutate(Different_utensil = replace(Different_utensil,
                                                            Different_utensil == "About Half the time",
                                                           "sometimes")) %>%
            mutate(Different_utensil = replace(Different_utensil,
                                                            Different_utensil == "Usually",
                                                           "seldom")) %>%
            mutate(Different_utensil = replace(Different_utensil,
                                                            Different_utensil == "Always",
                                                           "never"))

data_main <- mutate(data_main, Different_cutting_boards = replace(Different_cutting_boards,
                                                            Different_cutting_boards == "Never",
                                                           "always")) %>%
            mutate(Different_cutting_boards = replace(Different_cutting_boards,
                                                            Different_cutting_boards == "Seldom",
                                                           "usually")) %>%
            mutate(Different_cutting_boards = replace(Different_cutting_boards,
                                                            Different_cutting_boards == "Sometimes",
                                                           "about half the time")) %>%
            mutate(Different_cutting_boards = replace(Different_cutting_boards,
                                                            Different_cutting_boards == "About Half the time",
                                                           "sometimes")) %>%
            mutate(Different_cutting_boards = replace(Different_cutting_boards,
                                                            Different_cutting_boards == "Usually",
                                                           "seldom")) %>%
            mutate(Different_cutting_boards = replace(Different_cutting_boards,
                                                            Different_cutting_boards == "Always",
                                                           "never"))


data_main <- mutate(data_main, Not_leaving_food_out = replace(Not_leaving_food_out,
                                                            Not_leaving_food_out == "Never",
                                                           "always")) %>%
            mutate(Not_leaving_food_out = replace(Not_leaving_food_out,
                                                            Not_leaving_food_out == "Seldom",
                                                           "usually")) %>%
            mutate(Not_leaving_food_out = replace(Not_leaving_food_out,
                                                            Not_leaving_food_out == "Sometimes",
                                                           "about half the time")) %>%
            mutate(Not_leaving_food_out = replace(Not_leaving_food_out,
                                                            Not_leaving_food_out == "About Half the time",
                                                           "sometimes")) %>%
            mutate(Not_leaving_food_out = replace(Not_leaving_food_out,
                                                            Not_leaving_food_out == "Usually",
                                                           "seldom")) %>%
            mutate(Not_leaving_food_out = replace(Not_leaving_food_out,
                                                            Not_leaving_food_out == "Always",
                                                           "never"))
data_main <- mutate_all(data_main, tolower)



                    
dim(data_main)
data_main
```

```r
data_no_wash_why <- select(data, Rank_Feel_I_Should_but_Dont:Rank_Not_Cultural_or_Religious_Custom)
# filter(!is.na(col1))

names(data_no_wash_why) <- gsub(x = names(data_no_wash_why), pattern = "\\Rank_", replacement = "")

colnames(data_no_wash_why) <- c("Feel_I_should_but_dont", "Meat_is_clean_enough", "Exprts_advise_against", 
                                "Preserve_texture_etc", "Not_cultural_or_religious_custom")

data_no_wash_why <- replace(data_no_wash_why, is.na(data_no_wash_why), 0) %>%
                    replace(data_no_wash_why == 1, 6) %>%
                    replace(data_no_wash_why == 2, 5) %>%
                    replace(data_no_wash_why == 3, 4) %>%
                    replace(data_no_wash_why == 4, 3) %>%
                    replace(data_no_wash_why == 5, 2) %>%       
                    replace(data_no_wash_why == 6, 1)

head(data_no_wash_why)
```

```r
colnames(data)[35] <- "Concerned_about_the_health_and_safety_of_the_meat"
colnames(data)[40] <- "Trying_to_improve_the_appearance_of_the_meat"

data_wash_why <- select(data, Concerned_about_the_health_and_safety_of_the_meat:Trying_to_improve_the_appearance_of_the_meat)

colnames(data_wash_why) <- c("Health_and_safety_of_meat", "Religious_practice", "Non_religious_practice", "Improve_taste/smell",
                             "Improve_texture", "Improve_appearance")

data_wash_why <- replace(data_wash_why, is.na(data_wash_why), 0) %>%
                    replace(data_wash_why == 1, 7) %>%
                    replace(data_wash_why == 2, 6) %>%
                    replace(data_wash_why == 3, 5) %>%
                    replace(data_wash_why == 5, 3) %>%       
                    replace(data_wash_why == 6, 2) %>%
                    replace(data_wash_why == 7, 1)


head(data_wash_why)
```

```r
colnames(data)[43] <- "Remove_Pathogens"
colnames(data)[50] <- "Remove_Meat_Juice"

# 8 columns

data_wash_what <- select(data, Remove_Pathogens:Remove_Meat_Juice)

colnames(data_wash_what) <- c("Remove_pathogens", "Remove_artificial_chemicals", "Remove_debris", "Remove_undesirable_flavors_or_odors", 
                             "Remove_blood", "Remove_slime", "Remove_fat", "Remove_meat_juice")

data_wash_what <- replace(data_wash_what, is.na(data_wash_what), "0") %>%
                    replace(data_wash_what == "Very Important", "1") %>%
                    replace(data_wash_what == "Important", "1") %>%
                    replace(data_wash_what == "Moderately Important", "0") %>%
                    replace(data_wash_what == "Of Little Importance","0") %>%
                    replace(data_wash_what == "Unimportant", "0")

data_wash_what <- mutate_if(data_wash_what, is.character,as.numeric)

head(data_wash_what)
```

```r
# write out all dataframes to csv files
write_csv(data_main, "data/data_main.csv")
write_csv(data_no_wash_why, "data/data_no_wash_why.csv")
write_csv(data_wash_why, "data/data_wash_why.csv")
write_csv(data_wash_what, "data/data_wash_what.csv")
```
